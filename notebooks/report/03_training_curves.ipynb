# 03 - Training Curves

## Mục tiêu
Vẽ các đường cong training (loss, mAP, v.v.) từ logs đã có, phục vụ minh hoạ trong báo cáo.


```python
import json
from pathlib import Path
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-darkgrid")
plt.rcParams["figure.figsize"] = (12, 8)

PROJECT_ROOT = Path("D:/multi_modal_robot_ai").resolve()
FIG_DIR = PROJECT_ROOT / "experiments" / "figures" / "evaluation"
LOG_DIR = PROJECT_ROOT / "experiments" / "logs" / "vision_training_logs"

print("FIG_DIR =", FIG_DIR, "exists=", FIG_DIR.exists())
print("LOG_DIR =", LOG_DIR, "exists=", LOG_DIR.exists())
```

```python
# Nếu đã có file training_curves.png trong figures, chỉ cần hiển thị lại

curve_img = FIG_DIR / "training_curves.png"
if curve_img.exists():
    import cv2

    img = cv2.imread(str(curve_img))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title("Training Curves (from saved figure)")
    plt.axis("off")
    plt.tight_layout()
    plt.show()
else:
    print("⚠️ Không tìm thấy training_curves.png. Có thể vẽ trực tiếp từ logs nếu cần.")
```
