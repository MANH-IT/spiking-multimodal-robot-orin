# 05 - So sánh SNN vs ANN

## Mục tiêu
So sánh các mô hình (SNN vs ANN) theo các tiêu chí: FPS, latency, power, mAP và hình minh hoạ kết quả.


```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

plt.style.use("seaborn-v0_8-darkgrid")
plt.rcParams["figure.figsize"] = (12, 8)

# Bảng so sánh demo (có thể chỉnh theo số liệu thực tế)
data = {
    "Model": ["ANN", "SNN"],
    "FPS": [18.5, 24.2],
    "Latency (ms)": [55.0, 38.0],
    "Power (W)": [35.0, 28.0],
    "mAP": [0.58, 0.61],
}

comp_df = pd.DataFrame(data)
comp_df
```

```python
# Vẽ biểu đồ cột so sánh mAP và FPS

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# mAP
axes[0].bar(comp_df["Model"], comp_df["mAP"], color=["#3498db", "#2ecc71"])
axes[0].set_title("So sánh mAP")
axes[0].set_ylim(0, 1.0)
for i, v in enumerate(comp_df["mAP"]):
    axes[0].text(i, v + 0.02, f"{v:.2f}", ha="center")

# FPS
axes[1].bar(comp_df["Model"], comp_df["FPS"], color=["#9b59b6", "#e67e22"])
axes[1].set_title("So sánh FPS")
for i, v in enumerate(comp_df["FPS"]):
    axes[1].text(i, v + 0.5, f"{v:.1f}", ha="center")

plt.tight_layout()
plt.show()
```

## Kết luận (gợi ý viết)

- Mô hình **SNN** cho FPS cao hơn và latency thấp hơn so với **ANN**, phù hợp triển khai trên Jetson.
- Mức tiêu thụ điện năng của SNN thấp hơn, phù hợp cho robot di động.
- mAP của SNN tương đương hoặc nhỉnh hơn ANN, chứng tỏ hiệu quả của kiến trúc đề xuất.

Có thể chỉnh lại số liệu trong bảng cho khớp với kết quả thực tế của bạn.
