# 04 - Robot Touchscreen Interface

## Mục tiêu
Tổng hợp ảnh chụp màn hình giao diện robot UTT: màn hình chính, khung chat, bản đồ, camera.


```python
from pathlib import Path
import matplotlib.pyplot as plt
import cv2

plt.style.use("seaborn-v0_8-darkgrid")
plt.rcParams["figure.figsize"] = (12, 8)

PROJECT_ROOT = Path("D:/multi_modal_robot_ai").resolve()
UI_CAPTURE_DIR = PROJECT_ROOT / "experiments" / "figures" / "ui"

print("UI_CAPTURE_DIR =", UI_CAPTURE_DIR, "exists=", UI_CAPTURE_DIR.exists())

if not UI_CAPTURE_DIR.exists():
    print("⚠️ Chưa có ảnh chụp giao diện thật. Có thể lưu ảnh chụp màn hình vào:")
    print("   ", UI_CAPTURE_DIR)
```

```python
# Hiển thị tất cả ảnh trong thư mục UI_CAPTURE_DIR (nếu có)

if UI_CAPTURE_DIR.exists():
    img_paths = sorted(list(UI_CAPTURE_DIR.glob("*.png"))) + sorted(list(UI_CAPTURE_DIR.glob("*.jpg")))
    print(f"Tìm thấy {len(img_paths)} ảnh giao diện.")
    for p in img_paths:
        img = cv2.imread(str(p))
        if img is None:
            continue
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.figure()
        plt.imshow(img)
        plt.title(p.name)
        plt.axis("off")
        plt.tight_layout()
        plt.show()
else:
    print("⚠️ Thư mục UI_CAPTURE_DIR chưa tồn tại.")
```
